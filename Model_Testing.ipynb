```python
import keras
```

```python
model =  keras.models.load_model("/content/drive/MyDrive/Music_Genre_Dataset/Music_Genre_10_CNN.h5")
```

```python
import json
import os
import math
import numpy as np
import librosa, librosa.display
import matplotlib.pyplot as plt
```

```python
JSON_PATH = "/content/drive/MyDrive/Music_Genre_Dataset/mfcc_data.json"
```

```python
def load_data(JSON_PATH):

    with open(JSON_PATH, "r") as f:
        data = json.load(f)

    # convert lists to numpy arrays
    X = np.array(data["mfcc"])
    y = np.array(data["labels"])
    mapping = data["mapping"]

    print("Data succesfully loaded!")

    return  X, y, mapping
```

```python
X, y, mapping = load_data(JSON_PATH)
```

```python
unknown = ["b","cl","co","d","h","j","m","p","re","ro"]
known = ["b","p",2,"ro","cl","m","d","h","j","re","co"]
```

```python
for i in range(0,9900,100):

  print(f"{i/100}. Real Genre: {y[i]} - {mapping[y[i]]}", end = " | ")
  x_to_test = X[i]
  x_to_test = x_to_test[np.newaxis, ...]
  prediction0 = model.predict(x_to_test, verbose = 0)
  # get index with max value
  predicted_index = np.argmax(prediction0, axis=1)
  y_hat = int(predicted_index)
  print(f"Predicted Genre: {y_hat} - {mapping[y_hat]}", end = " | ")
  if y_hat == y[i]:
    print("--Correct Prediction--")
  else:
    print("--Incorrect Prediction--")
```

```python
correct = 0


for a,b in zip(X,y):

  print(f"Real Genre: {b} - {mapping[b]}", end = " | ")
  x_to_test = a[np.newaxis, ...]
  prediction = model.predict(x_to_test, verbose = 0)

  # get index with max value
  predicted_index = np.argmax(prediction, axis=1)
  y_hat = int(predicted_index)
  print(f"Predicted Genre: {y_hat} - {mapping[y_hat]}", end = " | ")

  if y_hat == b:
    correct += 1
    print("--Correct Prediction--")
  else:
    print("--Incorrect Prediction--")
```
